```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 random points from N(0, 1)
points = np.random.randn(1000)
print (len(points))

# Create a figure with two subplots side by side
fig, (ax_scatter, ax_hist) = plt.subplots(1, 2, gridspec_kw={'width_ratios': [3, 1]}, figsize=(8, 6), sharey=True)

# Scatter plot: random points vs their index (x = 0 for vertical alignment)
ax_scatter.plot(np.arange(0,len(points)),points, '.', alpha=0.5)
ax_scatter.set_xlabel("Random Points")
ax_scatter.set_ylabel("Value (N(0,1))")
# ax_scatter.set_xlim(-0.1, 0.1)  # narrow x-range to mimic vertical axis

# Horizontal histogram
ax_hist.hist(points, bins=30, orientation='horizontal', color='gray', alpha=0.7)
ax_hist.set_xlabel("Count")
# ax_hist.set_yticks([])  # remove redundant y ticks if desired

plt.tight_layout()
# plt.show()
plt.savefig("scatter_histogram.png", dpi=300, bbox_inches='tight')
```
